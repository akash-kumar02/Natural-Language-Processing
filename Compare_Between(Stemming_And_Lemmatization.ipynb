```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
words = ['charging' , 'charges', 'charged', 'study', 'studying', 'easliy', 'fairly','losses', 'was' , 'neccessary ', 'implementation ', 'permission', 'learn']
```

```python
for word in words:
  print(word,": ",ps.stem(word))
```

```python
sentence="Police is chasing the suspect as it ran away".split()
sentence
```

```python
for t in sentence:
  print(ps.stem(t))
```

```python
from nltk.stem.snowball import SnowballStemmer
snow_stemmer=SnowballStemmer(language='english')
```

```python
for word in words:
  print(word,": ",snow_stemmer.stem(word))
```

```python
for token in sentence:
  print(snow_stemmer.stem(token))
```

# **Lemmatization with nltk**

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import wordnet # import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("studying")
```

```python
lemmatizer.lemmatize("studying", pos=wordnet.VERB)
```

```python
lemmatizer.lemmatize("went", pos=wordnet.VERB)
```
