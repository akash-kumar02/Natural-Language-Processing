**Using SpaCy:**

```python
import spacy
# Load English model
nlp = spacy.load("en_core_web_sm")
```

**Load PDF**

```python
!pip install PyPDF2
import PyPDF2
```

```python
reader=PyPDF2.PdfReader("/content/NLP_PDF.pdf")
text=""
for page in reader.pages:
    text+=page.extract_text()
```

```python
text
```

**Tokenize Text**

```python
doc = nlp(text)
```

```python
doc
```

```python
import nltk
from nltk import word_tokenize
```

```python
nltk.download("punkt_tab") # it is pretrained model for tokenization
```

```python
for token in word_tokenize(text):
  print(token)
```

**Tokenize With Spacy**

```python
for token in doc:
    print(token.text)
```

**POS Tagging**

```python
for token in doc:
    print(token.text, token.pos_, token.tag_)
```

**Dependency / Syntactic Parsing**

```python
for token in doc:
    print(f"{token.text} --> {token.dep_} --> {token.head.text}")
```
