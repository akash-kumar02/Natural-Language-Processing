```python
import spacy
from spacy.matcher import Matcher
```

```python
# 1️. Load the English NLP model
nlp = spacy.load("en_core_web_sm")
```

```python
# 2️. Read the text file and create a Doc object.
with open("owlcreek.txt","r") as f:
  text=nlp(f.read())
```

```python
doc=nlp(text)
```

```python
#  3️. Count total tokens.
len(doc)
```

```python
# 4. Count total sentences.
sent=list(doc.sents)
len(sent)
```

```python
# 5️. Print the second sentence.
second_sent=sent[1]
second_sent
```

```python
# 6. Print token details for the second sentence.
for token in second_sent:
 print(f'{token.text:{10}} {token.pos_:{8}} {token.tag_:{6}} {spacy.explain(token.tag_)}')
```

```python
# 7️. Create a matcher to find "swimming vigorously".
matcher=Matcher(nlp.vocab)
pattern=[{'LOWER':'swimming'},{'IS_SPACE':True,'OP':'*'},{'LOWER':'vigorously'}]
matcher.add('Swimming', [pattern])
```

```python
matches = matcher(doc)
matches
```
