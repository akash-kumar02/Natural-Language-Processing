# **Pattern**

```python
text="My telephone number is 408-555-1234"
```

```python
phone=re.search(r'\d\d\d-\d\d\d\-\d\d\d\d',text)
```

```python
phone.group()
```

# **Groups**

```python
phone_pattern=re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
results=re.search(phone_pattern,text)
```

```python
results.group()
```

# **The wildcard charecter**

```python
re.findall(r".at","the cat in the in the hat sat here")
```

```python
re.findall(r"...at","the bat went splat")
```

```python
re.findall(r"\S+at","the bat went splat")
```

```python
# ends with a number
re.findall(r'\d$','this ends with a number2')
```

```python
# starts with a number
re.findall(r'^\d','1 is the loneliest number.')
```

# **Exclusion**

```python
# to exclude we can use the ^ symbol in conjunction with a set of brackets []
# anything inside the bracket exclude
```

```python
phrase="there are 3 nmbers 34 inside 5 this sentence."
re.findall('[^\d]',phrase)
```

```python
re.findall(r'[^\d]+',phrase)
```

```python
test_phrase = 'This is a string! But it has punctuation. How can we remove it?'
```

```python
re.findall('[^!.? ]+',test_phrase)
```

```python
clean=' '.join(re.findall('[^!.? ]+',test_phrase))
clean
```
