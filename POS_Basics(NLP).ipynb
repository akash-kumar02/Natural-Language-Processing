```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(u"The quick brown fox jumped over the lazy dog's back")
```

```python
print(doc[4].text,doc[4].pos_,doc[4].tag_,spacy.explain(doc[4].tag_))
```

```python
for token in doc:
    print(f'{token.text:{10}} {token.pos_:{8}} {token.tag_:{6}} {spacy.explain(token.tag_)}')
```

# **Counting POS Tags**

```python
doc = nlp(u"The quick brown fox jumped over the lazy dog's back")
pos_counts = doc.count_by(spacy.attrs.POS)
pos_counts
```

```python
doc.vocab[84].text
```

```python
for k,v in sorted(pos_counts.items()):
    print(f'{k}. {doc.vocab[k].text:{5}}: {v}')
```

```python
# count the different fine-grained tags:
TAG_counts = doc.count_by(spacy.attrs.TAG)

for k,v in sorted(TAG_counts.items()):
    print(f'{k}. {doc.vocab[k].text:{4}}: {v}')
```

```python
DEP_counts = doc.count_by(spacy.attrs.DEP)

for k,v in sorted(DEP_counts.items()):
    print(f'{k}. {doc.vocab[k].text:{4}}: {v}')
```

# **Visualizaing POS**

```python
from spacy import displacy
```

```python
doc = nlp(u"The quick brown fox jumped over the lazy dog's back")
displacy.render(doc,style='dep',jupyter=True,options={'distance':100})
```

```python
for token in doc:
    print(f'{token.text:{10}} {token.pos_:{7}} {token.dep_:{7}} {spacy.explain(token.dep_)}')
```

```python
displacy.serve(doc,style='dep',options={'distance':100})
```

```python
doc1=nlp(u"This is a sentence. This is another sentence, possibly longer than the other.")
spans=list(doc1.sents)
```

```python
options={'distance':110,'compact':'True','color':'white','bg':'#09a3d5','font':'Times'}
displacy.render(spans,style='dep',options=options)
```
