**Subword Tokenization**

```python
string="I am learning tokenization"
```

```python
new_string=string.split()
new_string
```

```python
list3=[]
for word in new_string:
  if word[-3:] == "ing":
    list3.append(word.partition("ing"))
  elif word[-7:] == "ization":
    list3.append(word.partition("ization"))
  else:
    list3.append(word)
print(list3)
```

```python
list4=[]
for element in list3:
  if type(element) is tuple:
    list4.append(element[0])
    list4.append(element[1])
  else:
    list4.append(element)
print(list4)
```

# **NER Tags---**

```python
doc8=nlp(u"Apple to build a Hong Kong factory for $6 million")
for token in doc8:
  print(token.text,end=' | ')
print('\n----')
for ent in doc8.ents:
  print(ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
```

```python
len(doc8.ents)
```

```python
d1=nlp(u"Apple is planning to acquire a startup in the United Kingdom for $1 billion. Tim Cook announced this in a press release from Cupertino. Meanwhile, Microsoft and Google are observing the deal closely.")
for token in d1:
  print(token.text,end=' | ')
print('\n----')
for ent in d1.ents:
  print(ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
```

```python
len(d1.ents)
```

## **Visualizing the dependency parse**

```python
doc=nlp(u"Over the last quarter Apple sold nearly 20thousand iPods for a profit of $6 million")
displacy.render(doc,style='ent',jupyter=True)
```
