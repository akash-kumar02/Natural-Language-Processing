Date **Formating**

```python
from datetime import datetime
today=datetime(year=2018,month=1,day=27)
print(f"{today}")
```

```python
from datetime import datetime
today=datetime(year=2018,month=1,day=27)
print(f"{today:%B %d, %Y}")
```

```python
from datetime import datetime
today=datetime(year=2018,month=1,day=27)
print(f"{today:%b %d, %Y}")
```

**Writing a file**

```python
# add a second argument to the function, 'w' which stands for write
# passing 'w+' lets us read and write to the file

my_file = open('test.text','w+')
```

```python
# write to the file
my_file.write('This is the new first line')
```

```python
my_file.seek(0)
my_file.read()
```

```python
my_file.close()
```

Appending to a **file**

```python
my_file = open('test.text','a+')
my_file.write('\nThis line is being appended to the test.text')
my_file.write('\nAnother line here')
```

```python
my_file.seek(0)
print(my_file.read())
```

```python
my_file.close()
```
